```python
# Drop non-numeric columns
features= data.drop(['ID','Diagnosis'], axis = 1)
```

```python
true_labels = data['Diagnosis'].map({'B': 0, 'M': 1})  # Convert 'Diagnosis' to numeric labels
```

```python
from sklearn.preprocessing import StandardScaler

# Scale and standardized the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Convert the scaled features back to a DataFrame
#scaled_data = pd.DataFrame(scaled_features, columns=features.columns)

# Display the first few rows of the scaled DataFrame
#print(scaled_data.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assume 'scaled_features' contains the scaled data
# Plot the first two features of the scaled data
plt.figure(figsize=(10, 8))
sns.scatterplot(x=scaled_features[:, 0], y=scaled_features[:, 1])
plt.title('Data Visualization Before Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
from sklearn.cluster import KMeans

# Initialize the KMeans algorithm
kmeans = KMeans(n_clusters=2, init='k-means++', random_state=42)

# Fit the KMeans algorithm to the scaled data
kmeans.fit(scaled_features)

# Add the cluster labels to the original DataFrame
data['cluster'] = kmeans.labels_

# Display the first few rows of the DataFrame with cluster labels
#print(data.head())
```

```python
#import matplotlib.pyplot as plt
#import seaborn as sns

# Display the cluster centers
cluster_centers = pd.DataFrame(scaler.inverse_transform(kmeans.cluster_centers_), columns=features.columns)
#print(cluster_centers)
print("Cluster Centers:\n", cluster_centers)
```

```python
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score
from scipy.optimize import linear_sum_assignment 
def calculate_clustering_accuracy(true_labels, cluster_labels):
    contingency_matrix = pd.crosstab(true_labels, cluster_labels)
    row_ind, col_ind = linear_sum_assignment(-contingency_matrix.values)
    return contingency_matrix.values[row_ind, col_ind].sum() / len(true_labels)

#accuracy = calculate_clustering_accuracy(true_labels, kmeans.labels_)
#print(f"Clustering Accuracy: {accuracy}")
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score

def evaluate_kmeans(init_method='k-means++', n_clusters=2, n_init=10):
    # Initialize KMeans
    kmeans = KMeans(n_clusters=n_clusters, init=init_method, n_init=n_init, random_state=42)
    
    # Fit the KMeans algorithm to the scaled data
    kmeans.fit(scaled_features)
    
    # Add the cluster labels to the original DataFrame
    data['cluster'] = kmeans.labels_
    
    # Retrieve centroids
    centroids = kmeans.cluster_centers_
    
    # Calculate accuracy
    accuracy = calculate_clustering_accuracy(true_labels, kmeans.labels_)
    
    # Calculate ARI and NMI if desired
    ari = adjusted_rand_score(true_labels, kmeans.labels_)
    nmi = normalized_mutual_info_score(true_labels, kmeans.labels_)
    
    # Print the results
    print(f"Initialization: {init_method}, Clusters: {n_clusters}, Init Attempts: {n_init}")
    print(f"Clustering Accuracy: {accuracy}")
    print(f"Centroids:\n{centroids}")
    print(f"Adjusted Rand Index (ARI): {ari}")
    print(f"Normalized Mutual Information (NMI): {nmi}")
    print("\n")
    
    return centroids
```

```python
# Example 1: Default KMeans with k-means++ and 2 clusters
evaluate_kmeans(init_method='k-means++', n_clusters=2, n_init=10)

# Example 2: KMeans with random initialization and 2 clusters
evaluate_kmeans(init_method='random', n_clusters=2, n_init=10)

# Example 3: KMeans with k-means++ and 3 clusters
evaluate_kmeans(init_method='k-means++', n_clusters=3, n_init=10)

# Example 4: KMeans with k-means++ and 2 clusters, but more initialization attempts
evaluate_kmeans(init_method='k-means++', n_clusters=2, n_init=20)
```

# SUPERVISED LEARNING

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Split the data
X_train, X_test, y_train, y_test = train_test_split(scaled_features, true_labels, test_size=0.2, random_state=42)

# Initialize and train the classifier
classifier = LogisticRegression(random_state=42)
classifier.fit(X_train, y_train)

# Make predictions
y_pred = classifier.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
# Confusion Matrix as a heatmap
from sklearn.metrics import confusion_matrix
conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

## **Results and Evaluation**

### **1. Initial Clustering:**
- **Clusters Evaluated:** The K-Means algorithm was tested with different numbers of clusters (`n_clusters`).
- **Accuracy Observed:**
  - With `n_clusters = 2`: The model achieved approximately **90% accuracy**.
  - With `n_clusters = 3`: The accuracy dropped to about **75%**, indicating that adding more clusters led to confusion between the categories.

### **2. Comparison with Class Labels:**
- **Cluster Assignment:** The clustering results were compared against the actual diagnosis labels (M for malignant and B for benign).
- **Observations:** The algorithm's ability to separate the data into meaningful groups was mixed, with some cases being misclassified due to overlapping feature values.

### **3. Exploration of Centroid Initialization:**
- **Impact of Initialization:** Different initial centroids were explored to see how they affected the accuracy of clustering. It was observed that the choice of initial centers significantly influenced the final clusters and their accuracy.
- **Attempts to Improve Accuracy:** Adjustments in the initialization strategy were made to improve clustering outcomes, but consistent accuracy gains were challenging to achieve.

### **4. Supervised Method Comparison:**
- A supervised learning method, likely a classification algorithm, was used as a benchmark to compare with the unsupervised K-Means results.
- **Supervised Method Performance:** The supervised method generally outperformed K-Means in terms of accuracy, highlighting the challenges of using unsupervised techniques for tasks where labeled data is available.

### **5. Conclusion:**
- **K-Means Limitations:** While K-Means provided some insight into data clustering, it struggled with accuracy compared to supervised methods, especially when multiple clusters were introduced.
- **Supervised Learning Advantage:** The comparison underscored the importance of using labeled data when available, as supervised methods typically offer better diagnostic performance.
